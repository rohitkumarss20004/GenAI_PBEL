# 2. Read daily temperature data from a CSV file using Pandas. Use NumPy to calculate the highest, lowest, and average temperatures.

```python
import numpy as np
import pandas as pd


# 1. Function to classify temperature categories
def classify_temperature(temp):
    if temp < 20:
        return "Cold"
    elif 20 <= temp <= 30:
        return "Moderate"
    else:
        return "Hot"


# 2. Read the daily temperature data from the CSV file using Pandas
# Replace 'temperature_data.csv' with your actual file path
try:
    df = pd.read_csv("/content/temperature.csv")

    # Convert the Temperature column to a NumPy array for calculations
    temp_array = df["Temperature"].to_numpy()

    # 3. Use NumPy to calculate highest, lowest, and average temperatures
    highest_temp = np.max(temp_array)
    lowest_temp = np.min(temp_array)
    average_temp = np.mean(temp_array)

    # Display the summary statistics
    print("--- Temperature Statistics ---")
    print(f"Highest Temperature: {highest_temp:.1f}°C")
    print(f"Lowest Temperature:  {lowest_temp:.1f}°C")
    print(f"Average Temperature: {average_temp:.1f}°C")
    print("-" * 30 + "\n")

    # 4. Apply the classification function to each day
    # We use pandas .apply() to run our function on every row of the column
    df["Category"] = df["Temperature"].apply(classify_temperature)

    # 5. Display the category for each day
    print("--- Daily Classification Results ---")
    print(df[["Date", "Temperature", "Category"]].to_string(index=False))

except FileNotFoundError:
    print(
        "Error: 'temperature_data.csv' not found. Please make sure the file path is correct."
    )
```
