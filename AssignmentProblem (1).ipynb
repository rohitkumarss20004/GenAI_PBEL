```python
# 1. Function to calculate HRA (20% of basic salary)
def calculate_hra(basic_salary):
    return 0.20 * basic_salary


# 2. Function to calculate DA (10% of basic salary)
def calculate_da(basic_salary):
    return 0.10 * basic_salary


# 3. Function to calculate the final net salary
def calculate_net_salary(basic_salary):
    hra = calculate_hra(basic_salary)
    da = calculate_da(basic_salary)
    net_salary = basic_salary + hra + da
    return hra, da, net_salary


# --- Main Program ---

# Accept inputs from the user
emp_name = input("Enter Employee Name: ")
try:
    basic = float(input("Enter Basic Salary: $"))

    # Perform calculations using functions
    hra_amount, da_amount, total_net_salary = calculate_net_salary(basic)

    # Display the salary breakdown
    print("\n" + "=" * 35)
    print(f" SALARY BREAKDOWN FOR: {emp_name.upper()}")
    print("=" * 35)
    print(f"Basic Salary:     ${basic:,.2f}")
    print(f"HRA (20%):        ${hra_amount:,.2f}")
    print(f"DA (10%):         ${da_amount:,.2f}")
    print("-" * 35)
    print(f"Net Salary:       ${total_net_salary:,.2f}")
    print("=" * 35)

except ValueError:
    print("Invalid numeric value entered for basic salary.")
```

```python
# 1. Function to calculate total marks
def calculate_total(marks_list):
    return sum(marks_list)


# 2. Function to calculate percentage
def calculate_percentage(total_marks, num_subjects):
    # Assumes each subject is out of 100 marks
    max_possible_marks = num_subjects * 100
    return (total_marks / max_possible_marks) * 100


# 3. Function to assign final grade based on percentage boundaries
def assign_grade(percentage):
    if percentage >= 90:
        return "A"
    elif percentage >= 75:
        return "B"
    elif percentage >= 60:
        return "C"
    elif percentage >= 40:
        return "D"
    else:
        return "F"


# --- Main Program ---

num_subjects = 5
student_marks = []

# Accept marks for 5 subjects from the user
print(f"Enter the marks for {num_subjects} subjects (out of 100):")
for i in range(1, num_subjects + 1):
    while True:
        try:
            mark = float(input(f"Subject {i} Marks: "))
            if 0 <= mark <= 100:
                student_marks.append(mark)
                break
            else:
                print("Please enter a valid mark between 0 and 100.")
        except ValueError:
            print("Invalid input. Please enter a number.")

# Perform calculations using functions
total = calculate_total(student_marks)
percentage = calculate_percentage(total, num_subjects)
grade = assign_grade(percentage)

# Display report card metrics
print("\n" + "=" * 30)
print("     STUDENT REPORT CARD     ")
print("=" * 30)
print(f"Total Marks:       {total:.1f} / {num_subjects * 100}")
print(f"Percentage:        {percentage:.2f}%")
print(f"Final Grade:       {grade}")
print("=" * 30)
```

```python
# 1. Store grade criteria in a tuple (as required, keeps boundaries immutable)
GRADE_CRITERIA = (
    (90, "A"),
    (75, "B"),
    (60, "C"),
    (40, "D"),
    (0, "F"),
)


# --- Functions ---


# Function to calculate total marks from a list
def calculate_total(marks):
    return sum(marks)


# Function to calculate percentage (assuming each subject is out of 100)
def calculate_percentage(total, num_subjects=5):
    return (total / (num_subjects * 100)) * 100


# Function using control statements (if-elif-else) to assign grades
def assign_grade(percentage):
    if percentage >= 90:
        return "A"
    elif percentage >= 75:
        return "B"
    elif percentage >= 60:
        return "C"
    elif percentage >= 40:
        return "D"
    else:
        return "F"


# Function to format and display a single student's details
def display_student_details(roll_no, details):
    print(f"Roll No:    {roll_no}")
    print(f"Name:       {details['Name']}")
    print(f"Marks:      {details['Marks']}")
    print(f"Total:      {details['Total_Marks']} / 500")
    print(f"Percentage: {details['Percentage']:.2f}%")
    print(f"Grade:      {details['Grade']}")
    print("-" * 40)


# --- Main Program ---

# Dictionary to store all student records
# Structure will be: { roll_no: {"Name": name, "Marks": [...], ...} }
student_records = {}

print("=== Student Performance Management System ===")

# Loop to input data for multiple students
while True:
    try:
        roll_no = input("\nEnter Roll Number (or type 'exit' to finish): ").strip()
        if roll_no.lower() == "exit":
            break

        if roll_no in student_records:
            print("This Roll Number already exists. Please use a unique one.")
            continue

        name = input("Enter Student Name: ").strip()

        # Inputting marks for 5 subjects and saving to a list
        marks_list = []
        print("Enter marks for 5 subjects (out of 100):")
        for i in range(1, 6):
            while True:
                try:
                    m = float(input(f"  Subject {i}: "))
                    if 0 <= m <= 100:
                        marks_list.append(m)
                        break
                    else:
                        print("  Please enter marks between 0 and 100.")
                except ValueError:
                    print("  Invalid input. Enter a valid number.")

        # Perform calculations via our functions
        total = calculate_total(marks_list)
        percentage = calculate_percentage(total)
        grade = assign_grade(percentage)

        # Storing all structured details into the inner dictionary
        student_records[roll_no] = {
            "Name": name,
            "Marks": marks_list,
            "Total_Marks": total,
            "Percentage": percentage,
            "Grade": grade,
        }

    except Exception as e:
        print(f"An error occurred: {e}. Let's try that student profile again.")

# Check if any records were actually added before printing reports
if student_records:
    # Loop to display all student records
    print("\n" + "=" * 40)
    print("         ALL STUDENT RECORDS            ")
    print("=" * 40)
    for roll, details in student_records.items():
        display_student_details(roll, details)

    # Logic to find and display the topper (highest percentage)
    topper_roll = None
    highest_pct = -1

    for roll, details in student_records.items():
        if details["Percentage"] > highest_pct:
            highest_pct = details["Percentage"]
            topper_roll = roll

    # Display the final topper details
    print("=" * 40)
    print("                TOPPER                  ")
    print("=" * 40)
    display_student_details(topper_roll, student_records[topper_roll])

else:
    print("\nNo student records were entered.")9
```
