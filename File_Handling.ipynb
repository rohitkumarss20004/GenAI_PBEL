# Q1. Student Pass/Fail Checker

```python
# --- Setup: Creating a sample file for testing ---
sample_data = "Amit,55\nPriya,42\nRahul,78\nSneha,35\nVikram,60"
with open("filename.txt", "w") as f:
    f.write(sample_data)
# ------------------------------------------------

# Initialize counters
passed_count = 0
failed_count = 0

print("--- Student Results ---")

# Open and read the file line by line
with open("filename.txt", "r") as file:
    for line in file:
        # Clean up whitespace and split name from marks
        line = line.strip()
        if not line:  # Skip empty lines
            continue

        name, marks_str = line.split(",")
        marks = int(marks_str)

        # check condition using if-else
        if marks > 45:
            print(f"{name}: Passed ({marks})")
            passed_count += 1
        else:
            print(f"{name}: Failed ({marks})")
            failed_count += 1

print("-" * 25)
print(f"Total Passed: {passed_count}")
print(f"Total Failed: {failed_count}")
```

# Q2. High-Salary Employee Filter


```python
# --- Setup: Creating a sample file for testing ---
sample_salaries = "Rajesh,45000\nAnanya,65000\nKaran,52000\nSunita,30000\nArjun,85000"
with open("salary.txt", "w") as f:
    f.write(sample_salaries)
# ------------------------------------------------

# Initialize counter
high_salary_count = 0

print("--- Employees with Salary > ₹50,000 ---")

# Open and read the file
with open("salary.txt", "r") as file:
    for line in file:
        line = line.strip()
        if not line:
            continue

        name, salary_str = line.split(",")
        salary = int(salary_str)

        # Single if condition to check threshold
        if salary > 50000:
            print(f"Name: {name} | Salary: ₹{salary:,}")
            high_salary_count += 1

print("-" * 40)
print(f"Total employees matching criteria: {high_salary_count}")
```

#Q3. Attendance Tracker

```python
# --- Setup: Creating a sample file for testing ---
sample_attendance = (
    "Rohan,Present\nNeha,Absent\nSimran,Present\nAman,Absent\nJyoti,Present"
)
with open("attendance.txt", "w") as f:
    f.write(sample_attendance)
# ------------------------------------------------

# Initialize counters
present_count = 0
absent_count = 0

print("--- List of Absent Students ---")

# Open and read the file
with open("attendance.txt", "r") as file:
    for line in file:
        line = line.strip()
        if not line:
            continue

        name, status = line.split(",")

        # Filter absent students and increment total counters
        if status.strip().lower() == "absent":
            print(f"- {name}")
            absent_count += 1
        else:
            present_count += 1

print("\n--- Attendance Summary ---")
print(f"Total Present: {present_count}")
print(f"Total Absent:  {absent_count}")
```
